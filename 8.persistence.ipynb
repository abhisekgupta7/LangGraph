```python
from langgraph.graph import StateGraph,START,END
from typing import TypedDict,Annotated,Literal
from langchain_ollama import ChatOllama
from langgraph.checkpoint.memory import InMemorySaver
```

```python
llm=ChatOllama(model="llama3.1:8b")
```

```python
class JokeState(TypedDict):
    topic: str
    joke: str
    explanation: str
```

```python
def generate_joke(state:JokeState):
    prompt=f"Generate a joke about {state['topic']}"
    response=llm.invoke(prompt).content
    return {'joke': response}
```

```python
def generate_explanation(state:JokeState):
    prompt=f"Write an explanation for this joke: {state['joke']}"
    response=llm.invoke(prompt).content
    return {'explanation': response}
```

```python
graph=StateGraph(JokeState)

graph.add_node('generate_joke',generate_joke)
graph.add_node('generate_explanation',generate_explanation)

graph.add_edge(START,'generate_joke')
graph.add_edge('generate_joke','generate_explanation')
graph.add_edge('generate_explanation',END)

checkpointer=InMemorySaver()
workflow=graph.compile(checkpointer=checkpointer)
```

```python
config1={"configurable":{"thread_id": "1"}}
workflow.invoke({'topic': 'chicken crossing the road'},config=config1)
```

```python
print(workflow.get_state(config1))
print(list(workflow.get_state_history(config1)))
```

```python
config2={"configurable":{"thread_id": "2"}}
workflow.invoke({'topic': 'programmers'},config=config2)
```

```python
print(workflow.get_state(config2))
print(list(workflow.get_state_history(config2)))
```
